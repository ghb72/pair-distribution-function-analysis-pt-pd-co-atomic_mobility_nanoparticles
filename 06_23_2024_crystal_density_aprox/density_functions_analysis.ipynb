## Introducción General
En el estudio de la materia condensada, desde líquidos amorfos hasta sólidos cristalinos, la caracterización de la estructura atómica es fundamental para comprender y predecir las propiedades de los materiales. Para ir más allá de la descripción idealizada de redes perfectas, los científicos emplean funciones estadísticas que revelan el orden local y de rango medio. Dos de las herramientas más poderosas y omnipresentes en este campo son la Función de Distribución de Pares (PDF) y la Función de Distribución Radial (RDF). Aunque a menudo se usan indistintamente en conversaciones informales, representan conceptos distintos con propósitos complementarios, y su correcta interpretación es crucial para un análisis riguroso.

La Función de Distribución Radial, comúnmente denotada como $g(r)$, es una herramienta teórica y computacional por excelencia. Responde a la pregunta: "Si nos situamos sobre un átomo promedio, ¿cómo varía la densidad de átomos circundantes en función de la distancia $r$?". La $g(r)$ es una función normalizada; su valor oscila alrededor de la unidad. Un valor mayor que uno a una distancia $r$ indica una mayor probabilidad de encontrar un par de átomos a esa separación en comparación con una distribución completamente aleatoria, señalando la presencia de una capa de coordinación. Por el contrario, un valor menor que uno implica una repulsión o una zona de exclusión. A medida que la distancia $r$ aumenta, la $g(r)$ converge a uno, reflejando la pérdida de correlación estructural y la transición hacia la densidad promedio del material a granel. Debido a esta convergencia, la RDF es ideal para visualizar el orden de corto alcance, como las distancias de enlace y las primeras capas de vecinos en líquidos, vidrios y defectos en cristales.

Por otro lado, la Función de Distribución de Pares, o $G(r)$, es la contraparte experimental, obtenida directamente a partir de experimentos de difracción de rayos X o neutrones. La PDF no es una función que oscila alrededor de la unidad, sino que sus oscilaciones se superponen a una línea base que decrece con la distancia. Esta función proporciona una medida directa de la densidad de pares atómicos, ponderada por el poder de dispersión de cada elemento, y revela no solo las distancias de enlace (picos agudos), sino también la coherencia estructural a distancias más largas. A diferencia de la RDF, cuyos picos se ensanchan y se desvanecen rápidamente, los picos en la PDF pueden persistir a distancias considerables en materiales con orden de rango medio, como los vidrios metálicos o polímeros semicristalinos. Este alcance extendido convierte a la PDF en la técnica preferida para estudiar nanoestructuras, materiales amorfos con topologías complejas y transiciones sutiles entre orden y desorden.

En esencia, mientras que la RDF es un concepto abstracto ideal para modelado computacional que destaca la estructura local relativa a la densidad media, la PDF es una cantidad medible que proporciona un mapa absoluto de las distancias interatómicas. Un físico de materiales recurre a la RDF en simulaciones de dinámica molecular para validar potenciales interatómicos o para entender la estructura de un líquido ideal. En el laboratorio, se apoya en la PDF para determinar la estructura de un nuevo nanomaterial sintetizado o para rastrear los cambios estructurales durante una transición de fase. Comprender las sinergias y diferencias entre estas dos funciones es, por tanto, indispensable para conectar el mundo del modelado atómico con la realidad experimental de la ciencia de materiales.

## Fundamentación Teórica

 La función de densidad de n-partículas $\rho^{n/N}(r_1, r_2, \ldots, r_n)$, que describe la probabilidad de encontrar $n$ partículas en posiciones $r_1, r_2, \ldots, r_n$, está definida como (Chandler, 1998):
$$\rho^{n/N}(r_1, r_2, \ldots, r_n) = \frac{N!}{(N-n)!} \frac{\int dr^{N-n} exp[-βU(rᴺ)]}{∫ drᴺ exp[-βU(rᴺ)]}$$


<!--    -->

obtenemos la función de correlación por pares $g(r)$, con $r=|r_1-r_2|$ como:
$$g(r) = g(r_1, r_2) = \frac{\rho^{2/N}(r_1, r_2)}{\rho^{2}}$$
donde $\rho$ es la densidad promedio de partículas. También descrita por:
$$\rho=\rho^{1/N}(r_1) = N/V$$
<!--  -->
Por otra parte se maneja la $G(r)$ total (intimamente relacionada a $h(r)$) en la literatura referente a estudios de difracción (Keen, 2001) es:
$$ G(r) = (\sum_{i=1}^{N} c_i \bar{b_i})^2 (g(r)-1)$$
donde $c_i$ es la proporción de especies $i$ en el material y $b_i$ es la media de su poder de dispersión.

En los estudios de difracción para el estudio de nanomateriales y materiales con baja coherencia estructural, en cambio se usa el formalismo "PDFFIT" de Proffen y Billinge (1999):
$$G^{PDF}(r) = 4 \pi r [\rho^{PDF}(r) - \rho_0]$$
donde $\rho^{PDF} = \rho_0 g(r)$ y 
$$g(r) = (\sum_{i=1}^{N} c_i \bar{b_i})^{-2} \sum_{i,j=1}^{N} c_i c_j \bar{b_i} \bar{b_j} g_{i,j} (r)$$
y para eso se define la función de correlación discreta $g_{i,j}(r)$ como:
$$g_{i,j}(r)= \frac{n_{i,j}(r)}{4\pi r^2 dr c_{j} \rho_{0}}$$
donde la densidad atómica promedio es $\rho_0 = N/V$ y $n_{i,j(r)}$ son el numero de partículas de tipo $j$ entre $r$ y $r + dr$ desde una partícula de tipo $i$.

### Valores límite

Las funciones $g(r)$ y $G^{PDF}(r)$ tienen los valores límites:
$$g(r<r^0) = 0$$
$$g(r\rightarrow \infty) = 1$$
$$G^{PDF}(r<r^0) \approx -4\pi r \rho_0$$
$$G^{PDF}(r\rightarrow \infty) = 0$$

### Relación entre $G^{PDF}(r)$ y $g(r)$
Por lo tanto tenemos que:

$$G^{PDF}(r) = 4 \pi \rho_0 r [g(r) - 1]$$
$$\rho^{PDF} \propto \rho_0 g(r)$$
$$g(r) \propto \frac{G^{PDF}(r)}{4 \pi r \rho_0} + 1$$

$$ G(r) =\frac{\rho^{PDF}(r)}{\rho_{0}} (\sum_{i=1}^{N} c_i \bar{b_i})^2$$

**Sobre las discrepancias en la PDF a bajo r:**

Experimental  
<small>
	La obtención de la Función de Distribución de Pares (PDF) a través de la transformada de Fourier de un patrón de difracción experimental (Jeong et al., 1999; Warren, 1990) introduce, de manera inevitable, artefactos que son más evidentes a valores bajos de r, específicamente en la región por debajo de la primera distancia interatómica. Estas discrepancias, que se manifiestan como oscilaciones no físicas o "rizos de terminación", no son un reflejo de la estructura real del material, sino una consecuencia matemática directa de las limitaciones del experimento. Una transformada de Fourier precisa, capaz de resolver perfectamente las posiciones atómicas, requeriría datos de difracción en un rango infinito del vector de dispersión Q. En la realidad, todo experimento está limitado a un valor máximo, Q<sub>max</sub>. Este corte abrupto en los datos de Q equivale a multiplicar la señal ideal por una función escalón, lo que, en el espacio real de r, resulta en la convolución de la PDF verdadera con una función instrumental que presenta oscilaciones. El efecto es particularmente notorio a bajo r porque la característica más abrupta de cualquier estructura atómica es la exclusión total de pares por debajo de la distancia de enlace más corta—un cambio de cero a un valor finito. Recrear esta discontinuidad tan aguda exigiría las frecuencias más altas (valores de Q muy grandes) que el experimento no puede capturar. En su ausencia, la transformada genera estos rizos espurios que se "filtran" en la región físicamente prohibida, creando una representación que, si bien es matemáticamente inevitable, debe ser reconocida como un artefacto y no como una característica estructural (Peterson et al., 2003).
</small>

Modelos Teóricos  
<small>
	Al comparar una Función de Distribución de Pares (PDF) calculada (Billinge & Thorpe, 2006; Billinge & Egami, 1993) a partir de un modelo estructural atómico con datos experimentales, a menudo surgen discrepancias significativas, especialmente a valores bajos de r. La raíz de esta diferencia no yace en un error del modelo, sino en su naturaleza idealizada. Un cálculo teórico, que parte de un conjunto discreto de coordenadas atómicas y sus correspondientes factores de dispersión, genera una PDF que es, por definición, idénticamente cero para cualquier distancia r menor que el enlace interatómico más corto. Esta es una representación perfecta de la estructura postulada. Sin embargo, este resultado "limpio" contrasta fuertemente con la PDF experimental que, como producto de una transformada de Fourier truncada, inevitablemente exhibe oscilaciones espurias en esta misma región físicamente prohibida. Por lo tanto, intentar reconciliar esta diferencia mediante simples normalizaciones o ajustes de escala —como los que se podrían usar para aproximar una Función de Distribución Radial (RDF)— es un enfoque conceptualmente erróneo (Olds et al., 2015). Tales manipulaciones pueden alterar la amplitud o la línea base de la PDF calculada, pero no pueden introducir los rizos de terminación o los límites para g(r<r<sub>0</sub>), ya que estos son un artefacto intrínseco del proceso de medición experimental en el espacio recíproco y no existen en el cálculo directo a partir del modelo. El verdadero desafío en el refinamiento no es forzar el modelo a tener oscilaciones no físicas, sino convolucionar la PDF teórica ideal con una función de resolución instrumental que simule adecuadamente el efecto de Q<sub>max</sub>, permitiendo así una comparación justa y cuantitativa con la señal experimental.
</small>

## Aproximación de la PDF a partir de la RDF
Se puede aproximar la PDF a partir de la RDF, sin embargo no es fiable la construcción de la PDF a partir de la RDF al introducir limitaciones teóricas y factores de normalización y escala que no son considerados en la RDF. Sin embargo, lo que mejor que pude hacer (hasta el momento) es la siguiente aproximación:

```python
import pandas as pd
import plotly.express as px
density = pd.read_csv('density.txt', sep='\\s+')
density.index = density['#r(A)']
density = density.drop(columns=['#r(A)'])
# px.line(density, title='g(r) obtenida por calculos de densidad para un material bulk')

plt.figure(figsize=(10, 4), layout='constrained')
plt.plot(density.index, density['g(r)_raw'], color = 'black', label = 'RDF g(r)')
plt.plot(density.index, density['Base'], color = 'blue', label = 'Base')
plt.legend()
plt.xlabel('r [$\\AA$]')
plt.ylabel('g(r)')
plt.title('RDF g(r) de un material bulk de Pt idealizado')
plt.show()
```

```python
mean_density = 0.0677058 # Densidad promedio del bulk
g_r = density.index * (density['g(r)_raw'] - 1)*mean_density
```

```python
px.line(g_r, title='PDF normalizada a partir de la RDF g(r)')
plt.figure(figsize=(10, 4), layout='constrained')
plt.plot(density.index,g_r, color = 'black', label = 'PDF g(r)')
plt.plot(density.index, (density['Base']-1)/density.index, color = 'blue', label = 'Base')
plt.xlabel('r [$\\AA$]')
plt.ylim(-0.25,0.25)
plt.ylabel('g(r)')
plt.title('PDF g(r) de una nanopartícula de Pt-Ni idealizada')
plt.legend()
plt.show()
```

## En conclusión
Aunque estén intimamente relacionadas, la PDF y la RDF son funciones diferentes que describen aspectos distintos de la estructura atómica. La PDF es una medida absoluta de la densidad de pares atómicos, mientras que la RDF es una función normalizada que describe la variación relativa de la densidad de átomos en función de la distancia. La comprensión de estas diferencias es crucial para el análisis preciso de datos experimentales y simulados en ciencia de materiales.

Los métodos de cálculo de ambas funciones son diferentes, y aunque la PDF puede aproximarse a partir de la RDF, esta aproximación no es exacta y puede introducir errores en el análisis. Por lo tanto, es importante utilizar las funciones adecuadas para cada tipo de análisis y comprender sus limitaciones y aplicaciones específicas.
