```python
from sys import path
import os
dir_actual = os.getcwd()
path.insert(0, 'C:/Users/guill/OneDrive/Documentos/Tesis/Proyect')

import myfuncions.core_shell_f as cs
import myfuncions.randomize_f as ranf
import myfuncions.tools_f as tools
import subprocess as sbp
import myfuncions.graphs_f as gf

a = [1,0.75,0.5,0.25,0,-0.25,-0.5,-0.75,-1]
b = [0,0.07,0.14,0.21,0.27001,0.44,0.64,0.83,1]
print(len(a), len(b))
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = [1,0.75,0.5,0.25,0.0,-0.25,-0.5,-0.75,-1]
b = [0,0.07,0.14,0.21,0.27001,0.44,0.64,0.83,1]
fig, ax = plt.subplots()
for i in range(0,1):
    aa = a[i]
    bb = b[i]
    x = np.arange(0,1,0.02)
    pol = aa*pow(x,3*aa/(3*0.27-bb)-3)+bb
    ax.plot(x,pol**2,label=f'M{i}')

ax.legend()
ax.grid()
ax.set_xlabel('Fracción de Radio')
ax.set_ylabel('Fracción de átomos de interés')
```
