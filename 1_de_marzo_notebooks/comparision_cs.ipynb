```python
import pandas as pd
import matplotlib.pyplot as plt

FCC = pd.read_csv('../15_de_marzo/Pt_100/Pt-FCC_d.txt', sep= '\s+')
Ni = pd.read_csv('../15_de_marzo/Ni_100/Ni-FCC_d.txt', sep= '\s+')
Nics50 = pd.read_csv('C:\\Users\\guill\\OneDrive\\Documentos\\Tesis\\Proyect\\1_de_marzo_notebooks\\NCsNi0.5\\Ni_0.5_cs_d.txt', sep = '\s+')
Nics60 = pd.read_csv('C:\\Users\\guill\\OneDrive\\Documentos\\Tesis\\Proyect\\1_de_marzo_notebooks\\NCsNi0.6\\Ni_0.6_cs_d.txt', sep = '\s+')

fig, ax =plt.subplots(figsize=(10,8), layout='constrained')
ax.plot(FCC['#r(A)'], FCC['G(r)']+0.4, color = 'black', label = '100% Pt')
ax.plot(Nics50['#r(A)'], Nics50['G(r)'], label = '50% Ni core, 50% Pt shell')
ax.plot(Nics60['#r(A)'], Nics60['G(r)']+0.2, label = '60% Ni core, 40% Pt shell')
ax.plot(Ni['#r(A)'], Ni['G(r)']-0.2, color = 'red', label = '100% Ni')


ax.set_xlabel('r [$\\AA$]')
ax.set_xlim([0,25])
ax.set_ylabel('G(r)')
#ax.set_title('Comparación de PDFs para nanopartículas janus de 3nm con \n diferententes proporciones de componentes')
#ax.grid('True', linestyle='-')
ax.legend()
plt.rc('font', size=12)
plt.rcParams['font.family'] = 'serif'
plt.show()
```
