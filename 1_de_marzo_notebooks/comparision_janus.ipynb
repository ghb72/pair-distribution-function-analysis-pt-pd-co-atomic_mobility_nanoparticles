```python
import pandas as pd
import matplotlib.pyplot as plt

FCC = pd.read_csv('../15_de_marzo/Pt_100/Pt-FCC_d.txt', sep= '\s+')
Ni = pd.read_csv('../15_de_marzo/Ni_100/Ni-FCC_d.txt', sep= '\s+')
j50 = pd.read_csv('C:\\Users\\guill\\OneDrive\\Documentos\\Tesis\\Proyect\\1_de_marzo_notebooks\\Janus_0.5\\janus-0.5_d.txt', sep = '\s+')
j25 = pd.read_csv('c:\\Users\\guill\\OneDrive\\Documentos\\Tesis\\Proyect\\1_de_marzo_notebooks\\Ni_0.75_janus\\Ni_0.75_janus_d.txt', sep = '\s+')
j75 = pd.read_csv('C:\\Users\\guill\\OneDrive\\Documentos\\Tesis\\Proyect\\1_de_marzo_notebooks\\Pt_0.75_janus\\Pt_0.75_janus_d.txt', sep = '\s+')

fig, ax =plt.subplots(figsize=(10,8), layout='constrained')
ax.plot(FCC['#r(A)'], FCC['G(r)']+0.9, color = 'black', label = '100% Pt')
ax.plot(j75['#r(A)'], j75['G(r)']+0.6, color = 'green', label = '75% Pt, 25% Ni')
ax.plot(j50['#r(A)'], j50['G(r)']+0.3, color = 'blue', label = '50% Pt, 50% Ni')
ax.plot(j25['#r(A)'], j25['G(r)'], color = 'red', label = '25% Pt, 75% Ni')
ax.plot(Ni['#r(A)'], Ni['G(r)']-0.3, color = 'orange', label = '100% Ni')


ax.set_xlabel('r [$\\AA$]')
ax.set_xlim([0,25])
ax.set_ylabel('G(r)')
#ax.set_title('Comparacin de PDFs para nanopartculas janus de 3nm con \n diferententes proporciones de componentes')
#ax.grid('True', linestyle='-')
ax.legend()
plt.rc('font', size=12)
plt.rcParams['font.family'] = 'serif'
plt.show()
```
