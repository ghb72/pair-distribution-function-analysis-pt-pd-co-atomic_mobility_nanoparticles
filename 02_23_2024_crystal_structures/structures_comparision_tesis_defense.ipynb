# COMPARACIÓN ENTRE LAS ESTRUCTURAS
Hago la comparación entre las estructuras FCC, BCC, HCP, e Icosaédrica

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


exp = pd.DataFrame()

df_exp = pd.read_csv('../04_26_2024_comparision_variables/Post-PtPdCo-PDF.csv', sep = '\s+', header=None, names=['r','G(r)'])
df_exp = df_exp.replace('--',np.nan).dropna()
rest_df_exp = df_exp["G(r)"].iloc[1::2].astype(float)
rest_df_exp = rest_df_exp.reset_index(drop=True)
rest_df_exp.index *= 0.02

exp['exp'] = rest_df_exp

dx = 0.02

BCC = pd.read_csv('C:/Users/guill/OneDrive/Documentos/Tesis/Proyect/02_23_2024_crystal_structures/BCC/BCC.txt', sep= '\s+')
FCC = pd.read_csv('C:/Users/guill/OneDrive/Documentos/Tesis/Proyect/02_23_2024_crystal_structures/FCC/FCC.txt', sep= '\s+')
HCP = pd.read_csv('C:/Users/guill/OneDrive/Documentos/Tesis/Proyect/02_23_2024_crystal_structures/HCP/HCP.txt', sep= '\s+')
ICO = pd.read_csv('C:/Users/guill/OneDrive/Documentos/Tesis/Proyect/02_23_2024_crystal_structures/ICO/ICO.txt', sep= '\s+')

fig, ax =plt.subplots(figsize=(10,5), layout='constrained')
ax.plot(BCC['#r(A)']*0.98, BCC['G(r)']+1.5, color = 'orange', label = 'BCC')
ax.plot(FCC['#r(A)']*0.98, FCC['G(r)']+0.5, color = 'red', label = 'FCC')
ax.plot(HCP['#r(A)']*0.98, HCP['G(r)']+1, color = 'blue', label = 'HCP')
for i in range(3):
	if i == 0:
		label = 'Experimental'
	else:
		label = None
	ax.plot(exp.index, exp['exp']+i*0.5+0.5, label=label, color='black')


ax.set_xlabel('r [$\\AA$]')
ax.set_xlim([0,30])
ax.set_ylabel('G(r)')
ax.set_title('Comparación de PDFs para nanopartículas de 3nm con diferentes estructuras')
#ax.grid('True', linestyle='-')
ax.legend(loc='upper right')
plt.rc('font', size=12)
plt.rcParams['font.family'] = 'serif'
plt.show()
```

Podemos notar como la estructura HCP al tener el mayor porcentaje de llenado presenta menos picos pronunciados en comparación con las otras estructuras (exceptuando la estructura icosaédrica), se puede apreciar que los picos que aparecen en su gran mayoría presentan anchos mayores que los picos presentes en otras estructuras y están presentados en gran cantidad, coorespondiendose con el hecho de que existen mayor cantidad de vecinos interatómicos que la teoría explica en comparación con otras estructuras.
